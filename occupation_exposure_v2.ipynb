### Find Overlap between ISCO task and AI task

```python
# find overlap subset matching

def find_subset_overlap(pattern_list, ai_task_combined):
    return any(any(item in ai_tuple for ai_tuple in ai_task_combined) for item in pattern_list)

isco_task['overlap_subset_webb'] = isco_task['pattern_webb'].apply(lambda x:[[pattern for pattern in sublist if find_subset_overlap(pattern, ai_task_webb)] for sublist in x])
isco_task['overlap_subset_johann'] = isco_task['pattern_johann'].apply(lambda x:[[pattern for pattern in sublist if find_subset_overlap(pattern, ai_task_johann)] for sublist in x])
isco_task['overlap_subset_siahaan'] = isco_task['pattern_siahaan'].apply(lambda x:[[pattern for pattern in sublist if find_subset_overlap(pattern, ai_task_siahaan)] for sublist in x])
isco_task['overlap_subset_new_pattern'] = isco_task['new_pattern'].apply(lambda x:[[pattern for pattern in sublist if find_subset_overlap(pattern, ai_task_new_pattern)] for sublist in x])
isco_task.head()
```

```python
# find overlap verb matching

def find_verb_overlap(isco_tasks, ai_tasks):
    row = []
    for isco_task in isco_tasks:
        task_list = []
        for isco_keyword in isco_task:
            for isco_item, isco_pos in isco_keyword:
                if isco_pos.startswith('VB'):
                    found_match = False
                    for ai_task in ai_tasks:
                        for ai_item_pos in ai_task:
                            if isco_item in ai_item_pos:
                                task_list.append(isco_keyword)
                                found_match = True
                                break
                        if found_match:
                            break
                if found_match:
                    break
        row.append(task_list)
    return row

isco_task['overlap_verb_webb'] = isco_task['pattern_webb'].apply(lambda x:find_verb_overlap(x, ai_task_webb))
isco_task['overlap_verb_johann'] = isco_task['pattern_johann'].apply(lambda x:find_verb_overlap(x, ai_task_johann))
isco_task['overlap_verb_siahaan'] = isco_task['pattern_siahaan'].apply(lambda x:find_verb_overlap(x, ai_task_siahaan))
isco_task['overlap_verb_new_pattern'] = isco_task['new_pattern'].apply(lambda x:find_verb_overlap(x, ai_task_new_pattern))
```

```python
# find overlap verb-noun matching

def find_verb_noun_overlap(isco_tasks, ai_tasks):
  row = []
  for isco_task in isco_tasks:
    task_list = []
    for isco_keyword in isco_task:
      found_match = False
      for ai_keyword in ai_tasks:
        for isco_item, isco_pos in isco_keyword:
          if isco_pos.startswith('VB'):
            for ai_item, ai_pos in ai_keyword:
              if isco_item == ai_item:
                for isco_item2, isco_pos2 in isco_keyword:
                  if isco_pos2.startswith('NN'):
                    for ai_item2, ai_pos2 in ai_keyword:
                      if isco_item2 == ai_item2:
                        task_list.append(isco_keyword)
                        found_match = True
                      if found_match:
                        break
                  if found_match:
                    break
              if found_match:
                break
          if found_match:
            break
        if found_match:
          break
    row.append(task_list)
  return row

# isco_task['overlap_verb_noun_webb'] = isco_task['pattern_webb'].apply(lambda x:find_verb_noun_overlap(x, ai_task_webb))
# isco_task['overlap_verb_noun_johann'] = isco_task['pattern_johann'].apply(lambda x:find_verb_noun_overlap(x, ai_task_johann))
isco_task['overlap_verb_noun_siahaan'] = isco_task['pattern_siahaan'].apply(lambda x:find_verb_noun_overlap(x, ai_task_siahaan))
isco_task['overlap_verb_noun_new_pattern'] = isco_task['new_pattern'].apply(lambda x:find_verb_noun_overlap(x, ai_task_new_pattern))
```

```python
# find overlap verb-noun matching 2

def find_verb_noun2_overlap(isco_tasks, ai_tasks):
  row = []
  for isco_task in isco_tasks:
    task_list = []
    for isco_keyword in isco_task:
      found_match = False
      for ai_keyword in ai_tasks:
        for isco_item, isco_pos in isco_keyword:
          if isco_pos.startswith('VB'):
            for ai_item, ai_pos in ai_keyword:
              if isco_item == ai_item and ai_pos.startswith('VB'):
                for isco_item2, isco_pos2 in isco_keyword:
                  if isco_pos2.startswith('NN'):
                    for ai_item2, ai_pos2 in ai_keyword:
                      if isco_item2 == ai_item2 and ai_pos2.startswith('NN'):
                        task_list.append(isco_keyword)
                        found_match = True
                      if found_match:
                        break
                  if found_match:
                    break
              if found_match:
                break
          if found_match:
            break
        if found_match:
          break
    row.append(task_list)
  return row

isco_task['overlap_verb_noun2_webb'] = isco_task['pattern_webb'].apply(lambda x:find_verb_noun2_overlap(x, ai_task_webb))
# isco_task['overlap_verb_noun2_johann'] = isco_task['pattern_johann'].apply(lambda x:find_verb_noun2_overlap(x, ai_task_johann))
# isco_task['overlap_verb_noun2_siahaan'] = isco_task['pattern_siahaan'].apply(lambda x:find_verb_noun2_overlap(x, ai_task_siahaan))
# isco_task['overlap_verb_noun2_new_pattern'] = isco_task['new_pattern'].apply(lambda x:find_verb_noun2_overlap(x, ai_task_new_pattern))
```

```python
# find overlap all pos matching

def are_lists_of_tuples_equal(list1, list2):
    if len(list1) != len(list2):
        return False
    for tuple1, tuple2 in zip(list1, list2):
        if tuple1 != tuple2:
            return False
    return True

def find_all_pos_overlap(isco_tasks, ai_tasks):
  row = []
  for isco_task in isco_tasks:
    task_list = []
    for isco_keyword in isco_task:
      found_match = False
      for ai_keyword in ai_tasks:
        if are_lists_of_tuples_equal(isco_keyword, ai_keyword):
          task_list.append(isco_keyword)
          found_match = True
        if found_match:
          break
    row.append(task_list)
  return row

isco_task['overlap_all_pos_webb'] = isco_task['pattern_webb'].apply(lambda x:find_all_pos_overlap(x, ai_task_webb))
isco_task['overlap_all_pos_johann'] = isco_task['pattern_johann'].apply(lambda x:find_all_pos_overlap(x, ai_task_johann))
isco_task['overlap_all_pos_siahaan'] = isco_task['pattern_siahaan'].apply(lambda x:find_all_pos_overlap(x, ai_task_siahaan))
isco_task['overlap_all_pos_new_pattern'] = isco_task['new_pattern'].apply(lambda x:find_all_pos_overlap(x, ai_task_new_pattern))
```

```python
# function for calculate exposure score

def count_exposure_task(row, overlap, pattern):
    all_task = sum(1 for sublist in row[pattern] if sublist)
    overlap_task = sum(1 for sublist in row[overlap] if sublist)
    try:
      return overlap_task / all_task
    except ZeroDivisionError:
      return 0
```
