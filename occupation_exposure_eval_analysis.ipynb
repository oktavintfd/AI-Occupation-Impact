```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('occupation_exposure_eval_data.xlsx')
data = data[["exposure_keyword_verb_noun2_webb", "exposure_keyword_verb_noun2_johann", "exposure_keyword_verb_noun2_siahaan", "exposure_keyword_verb_noun2_new_pattern", "exposure_average"]]
data_sorted = data.sort_values(by='exposure_average').reset_index(drop=True)
data_sorted
```

```python
# Plot the data
plt.figure(figsize=(10, 6))

# Plot exposure_azka and exposure_talitha in red
plt.scatter(data_sorted.index, data_sorted['exposure_keyword_verb_noun2_webb'], color='red', label='exposure webb')
plt.scatter(data_sorted.index, data_sorted['exposure_keyword_verb_noun2_johann'], color='blue', label='exposure johann')
plt.scatter(data_sorted.index, data_sorted['exposure_keyword_verb_noun2_siahaan'], color='green', label='exposure siahaan')
plt.scatter(data_sorted.index, data_sorted['exposure_keyword_verb_noun2_new_pattern'], color='orange', label='exposure new pattern')
plt.scatter(data_sorted.index, data_sorted['exposure_average'], color='black', label='exposure average')

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Exposure Value')
plt.title('Comparison of Exposure Values')
plt.legend()

# Show the plot
plt.show()
```

```python
# Keyword-based verb-noun2 match
exposure_data = data[["exposure_keyword_verb_noun2_webb", "exposure_keyword_verb_noun2_johann", "exposure_keyword_verb_noun2_siahaan", "exposure_keyword_verb_noun2_new_pattern", "exposure_average"]]
label_mapping = {
    "exposure_keyword_verb_noun2_webb": "Webb (2019)",
    "exposure_keyword_verb_noun2_johann": "Johann et al (2017)",
    "exposure_keyword_verb_noun2_siahaan": "Siahaan et al (2023)",
    "exposure_keyword_verb_noun2_new_pattern": "Proposed method",
    "exposure_average": "Ground Truth"
}
# Plotting with Matplotlib and Seaborn
plt.figure(figsize=(10, 6))
for column in exposure_data.columns:
    if column == 'exposure_average':
        sns.histplot(exposure_data[column], kde=True, label=label_mapping[column], color='black', linewidth=2, alpha=1.0, kde_kws={'linewidth': 3})
    else:
        sns.histplot(exposure_data[column], kde=True, label=label_mapping[column], alpha=0.5)

# Adding labels and title
plt.title('')
plt.xlabel('Exposure Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
