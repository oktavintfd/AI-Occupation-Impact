Import Libraries

```python
import pandas as pd
```

Preprocessing

```python
# read data patent

patent_task = pd.read_excel('export_result/paten_3pattern.xlsx')
patent_task['pattern_raharjana'] = patent_task['pattern_raharjana'].apply(lambda x:list(eval(str(x))))
patent_task.head()
```

```python
# read data isco

isco_task = pd.read_excel('export_result/isco_3pattern.xlsx')
isco_task['pattern_raharjana'] = isco_task['pattern_raharjana'].apply(lambda x:list(eval(str(x))))
isco_task.head()
```

```python
# get unique list of AI task

ai_task = patent_task['pattern_raharjana'].apply(pd.Series).stack().tolist()
unique_tuples = set(ai_task)
ai_task = list(unique_tuples)
print(ai_task)
```

Measure Overlap between ISCO task and AI task

```python

# find overlap

def find_overlap(text):
    overlap = [i for i in text if i in ai_task]
    return overlap

isco_task['overlap'] = isco_task['pattern_raharjana'].apply(lambda x:find_overlap(x))
isco_task.head()
```

```python
# check missing value

print(isco_task.isna().sum())
```

```python
# count exposure score

def count_exposure(row):
    return len(row['overlap']) / len(row['pattern_raharjana'])

# Menggunakan apply pada dataframe
isco_task['exposure'] = isco_task.apply(count_exposure, axis=1)

print(isco_task)
```

```python
# 5 occupation most impacted by AI 

sorted_isco_task = isco_task.sort_values('exposure', ascending=False)
sorted_isco_task.head()
```
