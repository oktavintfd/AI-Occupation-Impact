### Exposure Score Proportion

```python
def show_exposure_proportion(data):
    # Define the categories
    bins = [0, 0.33, 0.66, 1]
    labels = ['0 - 0.33', '>0.33 - 0.66', '>0.66 - 1']

    # Categorize the data
    category = pd.cut(data, bins=bins, labels=labels, include_lowest=True, right=False)

    # Count the number of values in each category
    category_counts = category.value_counts().sort_index()

    # Create the pie chart
    colors = plt.cm.Blues_r(np.linspace(0.2, 1, 3))
    plt.figure(figsize=(8, 8))
    plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
    plt.title('Distribution of Values in Categories')
    plt.show()
```

```python
show_exposure_proportion(data["exposure_keyword_verb_noun2_webb"])
```

```python
show_exposure_proportion(data["exposure_keyword_verb_noun2_johann"])
```

### Major Group Classification

```python
# # Extract the first digit of ISCO 08 Code
data['First_digit'] = data['ISCO 08 Code'].str[0]
```

```python
# Exposure new_pattern

# Group and sort
mean_exposure = data.groupby('First_digit')['exposure_keyword_verb_noun2_new_pattern'].mean()
groups = mean_exposure.index.tolist()

x_labels = {
    '1': 'Managers',
    '2': 'Professionals',
    '3': 'Technicians and Associate Professionals',
    '4': 'Clerical Support Workers',
    '5': 'Service and Sales Workers',
    '6': 'Skilled Agricultural, Forestry and Fishery Workers',
    '7': 'Craft and Related Trades Workers',
    '8': 'Plant and Machine Operators, and Assemblers',
    '9': 'Elementary Occupations'
}

# Create a boxplot for each group
plt.figure(figsize=(12, 8))
boxplot = sns.boxplot(x='First_digit', y='exposure_keyword_verb_noun2_new_pattern', data=data)

# Set x-axis labels to be numeric, since we'll use a legend for the actual labels
boxplot.set_xticklabels([str(i) for i in groups])

# Add a legend with the actual labels
handles = [plt.Line2D([0], [0], color='w', label=f'{i}: {x_labels[str(i)]}', marker='s', markersize=10, linestyle='') for i in groups]
plt.legend(handles=handles, title='Major Group', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding labels and title
plt.title('Exposure by Major Group')
plt.xlabel('Major Group')
plt.ylabel('Exposure')
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```

```python
# Melt the DataFrame to long format
melted_data = pd.melt(data, id_vars=['First_digit'],
                      value_vars=['exposure_keyword_verb_noun2_webb',
                                  'exposure_keyword_verb_noun2_johann',
                                  'exposure_keyword_verb_noun2_siahaan',
                                  'exposure_keyword_verb_noun2_new_pattern'],
                      var_name='method', value_name='exposure')

# Define the labels for the x-axis
x_labels = {
    '1': 'Managers',
    '2': 'Professionals',
    '3': 'Technicians and Associate Professionals',
    '4': 'Clerical Support Workers',
    '5': 'Service and Sales Workers',
    '6': 'Skilled Agricultural, Forestry and Fishery Workers',
    '7': 'Craft and Related Trades Workers',
    '8': 'Plant and Machine Operators, and Assemblers',
    '9': 'Elementary Occupations'
}

method_labels = {
    "exposure_keyword_verb_noun2_webb": "Webb (2019)",
    "exposure_keyword_verb_noun2_johann": "Johann et al (2017)",
    "exposure_keyword_verb_noun2_siahaan": "Siahaan et al (2023)",
    "exposure_keyword_verb_noun2_new_pattern": "Proposed method"
}

# Create a boxplot for each group and method
plt.figure(figsize=(15, 12))
boxplot = sns.boxplot(x='First_digit', y='exposure', hue='method', data=melted_data)

# Set x-axis labels to be numeric
boxplot.set_xticklabels([str(i+1) for i in boxplot.get_xticks()])

# Legend for methods
handles, labels = boxplot.get_legend_handles_labels()
method_labels_mapped = [method_labels[label] for label in labels]  # Map method labels
method_legend = plt.legend(handles, method_labels_mapped, title='Method', bbox_to_anchor=(0.05, -0.1), loc='upper center', ncol=1)
plt.gca().add_artist(method_legend)

# Add a legend with the actual labels for Major Groups
handles = [plt.Line2D([0], [0], color='w', label=f'{i}: {x_labels[str(i)]}', marker='s', markersize=10, linestyle='') for i in x_labels.keys()]
plt.legend(handles=handles, title='Major Group', bbox_to_anchor=(0.4, -0.1), loc='upper center', ncol=2)

# Adding labels and title
plt.title('')
plt.xlabel('Major Group')
plt.ylabel('Exposure')
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```
