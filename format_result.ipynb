```python
import pandas as pd
```

```python
data = pd.read_excel('occupation_exposure_result_final.xlsx')
data['pattern_webb'] = data['pattern_webb'].apply(lambda x:list(eval(str(x))))
data['pattern_johann'] = data['pattern_johann'].apply(lambda x:list(eval(str(x))))
data['pattern_siahaan'] = data['pattern_siahaan'].apply(lambda x:list(eval(str(x))))
data['new_pattern'] = data['new_pattern'].apply(lambda x:list(eval(str(x))))
data['overlap_subset_webb'] = data['overlap_subset_webb'].apply(lambda x:list(eval(str(x))))
data['overlap_subset_johann'] = data['overlap_subset_johann'].apply(lambda x:list(eval(str(x))))
data['overlap_subset_siahaan'] = data['overlap_subset_siahaan'].apply(lambda x:list(eval(str(x))))
data['overlap_subset_new_pattern'] = data['overlap_subset_new_pattern'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_webb'] = data['overlap_verb_webb'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_johann'] = data['overlap_verb_johann'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_siahaan'] = data['overlap_verb_siahaan'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_new_pattern'] = data['overlap_verb_new_pattern'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_noun_webb'] = data['overlap_verb_noun_webb'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_noun_johann'] = data['overlap_verb_noun_johann'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_noun_siahaan'] = data['overlap_verb_noun_siahaan'].apply(lambda x:list(eval(str(x))))
data['overlap_verb_noun_new_pattern'] = data['overlap_verb_noun_new_pattern'].apply(lambda x:list(eval(str(x))))
data['overlap_all_pos_webb'] = data['overlap_all_pos_webb'].apply(lambda x:list(eval(str(x))))
data['overlap_all_pos_johann'] = data['overlap_all_pos_johann'].apply(lambda x:list(eval(str(x))))
data['overlap_all_pos_siahaan'] = data['overlap_all_pos_siahaan'].apply(lambda x:list(eval(str(x))))
data['overlap_all_pos_new_pattern'] = data['overlap_all_pos_new_pattern'].apply(lambda x:list(eval(str(x))))
data.head()
```

```python
import re

def split_task(row):
  result = re.split(r'\n', row)
  return result

data['Tasks include'] = data['Tasks include'].apply(lambda x: split_task(x))
data.head()
```

```python
# Columns with lists
list_columns = ['Tasks include', 'pattern_webb', 'pattern_johann', 'pattern_siahaan', 'new_pattern', 'overlap_subset_webb',	'overlap_subset_johann',	'overlap_subset_siahaan',	'overlap_subset_new_pattern', 'overlap_verb_webb',	'overlap_verb_johann',	'overlap_verb_siahaan',	'overlap_verb_new_pattern', 'overlap_verb_noun_webb',	'overlap_verb_noun_johann',	'overlap_verb_noun_siahaan',	'overlap_verb_noun_new_pattern', 'overlap_all_pos_webb',	'overlap_all_pos_johann',	'overlap_all_pos_siahaan',	'overlap_all_pos_new_pattern']

# Columns without lists
non_list_columns = [col for col in data.columns if col not in list_columns]

# Function to repeat non-list columns and expand list columns
def explode_lists(df, list_columns, non_list_columns):
    exploded_df = df[list_columns].apply(pd.Series.explode)
    repeated_non_list_df = df[non_list_columns].loc[df.index.repeat(df[list_columns[0]].str.len())].reset_index(drop=True)
    expanded_df = pd.concat([repeated_non_list_df, exploded_df.reset_index(drop=True)], axis=1)
    return expanded_df

# Apply the function to explode the lists
result_data = explode_lists(data, list_columns, non_list_columns)

# Display the result
print(result_data)
```

```python
result_data.to_excel('occupation_exposure_result_final_formatted.xlsx')
```
